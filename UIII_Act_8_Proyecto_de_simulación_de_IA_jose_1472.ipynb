### TORERO Y CAPOTE

```python
# Inicialización de variables de estado
energia = 100
fama = 0
capote_equipado = "Ninguno"

print("--- BIENVENIDO AL SIMULADOR DE TORERO IA 'EL ARTE DEL CAPOTE' ---")
print("Tu objetivo: Lograr la gloria en el ruedo manejando tu energía.")

# Bucle infinito para mantener la interactividad [1, 2]
while True:
    print("\n" + "-"*40)
    print(f"ESTADO: Energía: {energia} | Fama: {fama} | Capote: {capote_equipado}")
    print("-"*40)
    print("1. Elegir Capote (Herramienta)")
    print("2. Realizar una Verónica (Básico)")
    print("3. Descansar y beber agua (Recuperar)")
    print("4. Realizar un Pase de Pecho (Riesgo Alto)") # Nueva opción
    print("5. Finalizar la faena (Salir)")

    opcion = input("\n¿Qué decides hacer, Torero? (1-5): ")

    # Opción 1: Selección de Capote [3]
    if opcion == "1":
        print("\nIA: Elige tu herramienta:")
        eleccion = input("Escribe 'Pesado' o 'Ligero': ")
        if eleccion == "Pesado" or eleccion == "pesado":
            capote_equipado = "Pesado"
            print("IA: Capote Pesado equipado. Genera más fama pero cansa más.")
        elif eleccion == "Ligero" or eleccion == "ligero":
            capote_equipado = "Ligero"
            print("IA: Capote Ligero equipado. Mayor agilidad, menos esfuerzo.")
        else:
            print("IA ERROR: Opción no válida.")

    # Opción 2: Maniobra Verónica [3, 4]
    elif opcion == "2":
        if capote_equipado == "Ninguno":
            print("IA AVISO: ¡Necesitas un capote para esta maniobra!")
        elif energia >= 20:
            print("\n¡Ejecutas una Verónica con elegancia!")
            if capote_equipado == "Pesado" and energia > 50:
                fama = fama + 30
                energia = energia - 40
                print("IA: ¡Público impresionado! +30 Fama.")
            else:
                fama = fama + 15
                energia = energia - 15
                print("IA: Pase limpio. +15 Fama.")
        else:
            print("IA ALERTA: Sin energía suficiente para torear.")

    # Opción 3: Recuperar energía [4]
    elif opcion == "3":
        print("\nIA: Descansando...")
        if energia < 100:
            energia = energia + 30
            if energia > 100:
                energia = 100
            print(f"IA: Energía recuperada a {energia}.")
        else:
            print("IA: Ya estás al máximo de energía.")

    # Opción 4: Pase de Pecho (Nueva Opción de Riesgo) [3, 4]
    elif opcion == "4":
        if capote_equipado == "Ninguno":
            print("IA AVISO: El toro no esperará, ¡ponte el capote!")
        elif energia >= 45: # Requiere más energía que la Verónica
            print("\n¡Te la juegas con un Pase de Pecho monumental!")
            # Lógica de recompensa por riesgo: Capote pesado + mucha energía [3]
            if capote_equipado == "Pesado" and energia >= 70:
                fama = fama + 60 # Gran recompensa
                energia = energia - 60 # Gran cansancio
                print("IA: ¡ESTO ES ARTE! La plaza se pone en pie. +60 Fama.")
            else:
                fama = fama + 25
                energia = energia - 35
                print("IA: Maniobra arriesgada completada con éxito. +25 Fama.")
        else:
            print("IA ALERTA: Estás demasiado débil para este pase. ¡Peligro de cornada!")

    # Opción 5: Salir del programa [5]
    elif opcion == "5":
        print("\n--- RESUMEN FINAL ---")
        print(f"Fama: {fama} | Energía final: {energia}")
        if fama >= 80:
            print("¡ERES UNA LEYENDA! Sales a hombros.")
        else:
            print("Buena faena, pero falta práctica para la gloria.")
        break # Detiene el ciclo While True

    else:
        print("IA ERROR: Esa acción no existe en el ruedo.")
```
